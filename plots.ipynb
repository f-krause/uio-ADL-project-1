# Notebook to generate plots

## Compare training prorgress of best models

```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams["figure.figsize"] = (7,4)
```

```python
arr_full = np.loadtxt("output/losses_full_model.csv", delimiter=",", dtype=float)
arr_lora = np.loadtxt("output/losses_lora_model.csv", delimiter=",", dtype=float)
```

```python
plt.figure(facecolor='white')
plt.plot(list(range(1, arr_full.shape[0]+1)), arr_full, label = "full fine-tuning", marker="o", linestyle="-")
plt.plot(list(range(1, arr_lora.shape[0]+1)), arr_lora, label = "LoRA", marker="o", linestyle="-")
plt.xlabel("epoch")
plt.ylabel("cross entropy loss")
plt.title("Convergence progress")
plt.legend()

#plt.savefig("resources/losses.png", bbox_inches='tight') # Save image as svg

plt.show()
```

## Compare training path for each r

```python
files = ["losses_lora_r8_model.csv", "losses_lora_r16_model.csv", "losses_lora_r64_model.csv"]
```

```python
plt.figure(facecolor='white')

for file in files:
    data = np.loadtxt(f"experiments/{file}", delimiter=",", dtype=float)
    r = file.split("_")[2][1:]
    plt.plot(list(range(1, data.shape[0]+1)), data, label = "r=" + str(r), marker="o", linestyle="-")

plt.xlabel("epoch")
plt.ylabel("cross entropy loss")
plt.title("Convergence progress")
plt.legend()

#plt.savefig("resources/losses_lora_tuning.png", bbox_inches='tight') # Save image as svg

plt.show()
```
