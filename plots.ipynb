# Notebook to generate plots

```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams["figure.figsize"] = (7,4)
```

```python
arr_full = np.loadtxt("output/losses_full_model_old.csv", delimiter=",", dtype=float)
arr_lora = np.loadtxt("output/losses_lora_model_old.csv", delimiter=",", dtype=float)
```

```python
plt.plot(list(range(1, arr_full.shape[0]+1)), arr_full, label = "full fine-tuning", marker="o", linestyle="-")
plt.plot(list(range(1, arr_lora.shape[0]+1)), arr_lora, label = "LoRA", marker="o", linestyle="-")
plt.xlabel("epoch")
plt.ylabel("cross entropy loss")
plt.title("Convergence progress")
plt.legend()

#plt.savefig("resources/losses.svg", bbox_inches='tight') # Save image as svg

plt.show()
```
